```python
import pandas as pd
```

```python
#leer datos icfes 2010-2021
results_df = pd.read_csv ('rows.csv',low_memory = False)
```

```python
#filtrar por departamento
data_valle = results_df[results_df['COLE_DEPTO_UBICACION']=='VALLE']
```

```python
#filtrar datos con puntajes no nulos
data_valle = data_valle[data_valle['PUNT_GLOBAL'].notnull()]
```

```python
#filtrar rango de periodo de presentacion de 20171 en adelante
data_valle = data_valle[data_valle['PERIODO']>=20171]
```

```python
#transformar periodo en dato string
data_valle = data_valle.astype({'PERIODO':'str'})
```

```python
#transformar periodos 20171-20172 -> 2017 ... 2021
for i in range(0, len(data_valle)):
    year=data_valle["PERIODO"].iloc[i]
    data_valle['PERIODO'].iloc[i] = year[0:-1]
    print(data_valle['PERIODO'].iloc[i])
```

```python
#eliminar columnas no utiles
data_valle.drop(['Unnamed: 0.1', 'Unnamed: 0','ESTU_CONSECUTIVO','COLE_CALENDARIO','COLE_COD_DANE_ESTABLECIMIENTO', 'COLE_COD_DANE_SEDE', 'COLE_COD_MCPIO_UBICACION', 'COLE_COD_DEPTO_UBICACION', 'COLE_CODIGO_ICFES', 'COLE_GENERO', 'COLE_MCPIO_UBICACION', 'COLE_NATURALEZA', 'COLE_NOMBRE_ESTABLECIMIENTO', 'COLE_NOMBRE_SEDE' , 'COLE_SEDE_PRINCIPAL', 'ESTU_COD_DEPTO_PRESENTACION', 'ESTU_COD_DEPTO_PRESENTACION', 'ESTU_COD_MCPIO_PRESENTACION', 'ESTU_COD_RESIDE_DEPTO', 'ESTU_COD_RESIDE_MCPIO', 'ESTU_DEPTO_PRESENTACION', 'ESTU_DEPTO_RESIDE', 'ESTU_ESTADOINVESTIGACION', 'ESTU_ESTUDIANTE', 'ESTU_MCPIO_PRESENTACION', 'ESTU_NACIONALIDAD', 'ESTU_PAIS_RESIDE', 'ESTU_PRIVADO_LIBERTAD', 'FAMI_CUARTOSHOGAR', 'FAMI_PERSONASHOGAR', 'FAMI_TIENEAUTOMOVIL', 'FAMI_TIENELAVADORA'], axis=1, inplace=True)
```

```python
#filtrar que los campos de las columnas utiles no sean nulas
data_valle = data_valle[data_valle['PERIODO'].notnull()]
data_valle = data_valle[data_valle['ESTU_TIPODOCUMENTO'].notnull()]
data_valle = data_valle[data_valle['COLE_AREA_UBICACION'].notnull()]
data_valle = data_valle[data_valle['COLE_CARACTER'].notnull()]
data_valle = data_valle[data_valle['COLE_DEPTO_UBICACION'].notnull()]
data_valle = data_valle[data_valle['COLE_JORNADA'].notnull()]
data_valle = data_valle[data_valle['ESTU_FECHANACIMIENTO'].notnull()]
data_valle = data_valle[data_valle['ESTU_GENERO'].notnull()]
data_valle = data_valle[data_valle['ESTU_MCPIO_RESIDE'].notnull()]
data_valle = data_valle[data_valle['FAMI_EDUCACIONMADRE'].notnull()]
data_valle = data_valle[data_valle['FAMI_EDUCACIONPADRE'].notnull()]
data_valle = data_valle[data_valle['FAMI_ESTRATOVIVIENDA'].notnull()]
data_valle = data_valle[data_valle['FAMI_TIENECOMPUTADOR'].notnull()]
data_valle = data_valle[data_valle['FAMI_TIENEINTERNET'].notnull()]
data_valle = data_valle[data_valle['DESEMP_INGLES'].notnull()]
data_valle = data_valle[data_valle['PUNT_INGLES'].notnull()]
data_valle = data_valle[data_valle['PUNT_MATEMATICAS'].notnull()]
data_valle = data_valle[data_valle['PUNT_SOCIALES_CIUDADANAS'].notnull()]
data_valle = data_valle[data_valle['PUNT_C_NATURALES'].notnull()]
data_valle = data_valle[data_valle['PUNT_LECTURA_CRITICA'].notnull()]
data_valle = data_valle[data_valle['PUNT_GLOBAL'].notnull()]
data_valle
```

```python
import datetime
```

```python
#obtiene posiciones de fecha de nacimiento no utiles las cuales empiezan el año en 0
arr_date_delete = []

for i in range(0, len(data_valle)):
    date_estu = data_valle["ESTU_FECHANACIMIENTO"].iloc[i]
    split = date_estu.split('/')
    year = int(split[2])
    #d/m/aa 
    if(split[2].startswith('0')):
        print(split[2])
        arr_date_delete.append(i)
```

```python
#elimina registos no validos a partir de las posiciones obtenidas anteriormente
for i in range(0, len(arr_date_delete)):
    pos = arr_date_delete[i]-i  
    print(pos)
    data_valle = data_valle.drop(data_valle.index[pos])
```

```python
arr_age_delete = []

#transformar fecha de nacimiento a edades y obtiene edades no utiles min 14 max 40
today = datetime.datetime.today()

for i in range(0, len(data_valle)):
    date_estu = data_valle["ESTU_FECHANACIMIENTO"].iloc[i]
    
    date_estu = date_estu.replace("/", "-")
    date_estu = datetime.datetime.strptime(date_estu, '%d-%m-%Y')
    
    ndays = (today - date_estu).days
    age = int(ndays/365)
    if(age > 40 or age < 14):
        arr_age_delete.append(i)
    else:
        data_valle["ESTU_FECHANACIMIENTO"].iloc[i] = str(age)
```

```python
#eliminar edades no validas min 14 max 40
for i in range(0, len(arr_age_delete)):
    pos = arr_age_delete[i]-i  
    print(pos)
    data_valle = data_valle.drop(data_valle.index[pos])
```

```python
data_valle
```

```python
data_valle.rename(columns = {'ESTU_FECHANACIMIENTO':'ESTU_EDAD'}, inplace = True)
```

```python
#transformar datos de colegio bilingue s -> Si , n -> No, y obtiene posicion de datos nan
arr_cole_delete = []

for i in range(0, len(data_valle)):
    cole_bilingue = data_valle["COLE_BILINGUE"].iloc[i]
    #print(cole_bilingue)
    if(cole_bilingue!='S' and cole_bilingue!='N'):
        arr_cole_delete.append(i)
    if(cole_bilingue=='N'):
        cole_bilingue = 'Si'
        data_valle["COLE_BILINGUE"].iloc[i] = cole_bilingue
    if(cole_bilingue=='N'):
        cole_bilingue = 'No'
        data_valle["COLE_BILINGUE"].iloc[i] = cole_bilingue
```

```python
len(arr_cole_delete)
```

```python
#eliminar datos cole_bilingue no validas -> nan
for i in range(0, len(arr_cole_delete)):
    pos = arr_cole_delete[i]-i  
    print(pos)
    data_valle = data_valle.drop(data_valle.index[pos])
```

```python
#eliminar columnas inncesarias
data_valle.drop(['Unnamed: 0','Unnamed: 0.1', 'Unnamed: 0.2',], axis=1, inplace=True)
```
