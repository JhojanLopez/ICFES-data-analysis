```python
import datetime
```

```python
#obtiene posiciones de fecha de nacimiento no utiles las cuales empiezan el año en 0
arr_date_delete = []

for i in range(0, len(data_valle)):
    date_estu = data_valle["ESTU_FECHANACIMIENTO"].iloc[i]
    split = date_estu.split('/')
    year = int(split[2])
    #d/m/aa 
    if(split[2].startswith('0')):
        print(split[2])
        arr_date_delete.append(i)
```

```python
#elimina registos no validos a partir de las posiciones obtenidas anteriormente
for i in range(0, len(arr_date_delete)):
    pos = arr_date_delete[i]-i  
    print(pos)
    data_valle = data_valle.drop(data_valle.index[pos])
```

```python
data_valle
```

```python
arr_age_delete = []

#transformar fecha de nacimiento a edades y obtiene edades no utiles min 14 max 40
for i in range(0, len(data_valle)):
    date_estu = data_valle["ESTU_FECHANACIMIENTO"].iloc[i]
    date_estu = date_estu.replace("/", "-")
    period = data_valle["PERIODO"].iloc[i]
    
    split = date_estu.split('-')
    date_presentation = split[0]+'-'+split[1]+'-'+str(period)
    
    try:
        date_presentation = datetime.datetime.strptime(date_presentation, '%d-%m-%Y')
    except:
        arr_age_delete.append(i)
        continue
        
        
    date_estu = datetime.datetime.strptime(date_estu, '%d-%m-%Y')
    
    ndays = (date_presentation - date_estu).days
    age = int(ndays/365)
    
    if(age > 40 or age < 14):
        arr_age_delete.append(i)
    else:
        data_valle["ESTU_FECHANACIMIENTO"].iloc[i] = str(age)
```

```python
#eliminar edades no validas min 14 max 40
for i in range(0, len(arr_age_delete)):
    pos = arr_age_delete[i]-i  
    print(pos)
    data_valle = data_valle.drop(data_valle.index[pos])
```

```python
data_valle.rename(columns = {'ESTU_FECHANACIMIENTO':'ESTU_EDAD'}, inplace = True)
```

```python
#transformar datos de colegio bilingue s -> Si , n -> No, y obtiene posicion de datos nan
arr_cole_delete = []

for i in range(0, len(data_valle)):
    cole_bilingue = data_valle["COLE_BILINGUE"].iloc[i]
    #print(cole_bilingue)
    if(cole_bilingue!='S' and cole_bilingue!='N'):
        arr_cole_delete.append(i)
    if(cole_bilingue=='S'):
        cole_bilingue = 'Si'
        data_valle["COLE_BILINGUE"].iloc[i] = cole_bilingue
    if(cole_bilingue=='N'):
        cole_bilingue = 'No'
        data_valle["COLE_BILINGUE"].iloc[i] = cole_bilingue
```

```python
#eliminar datos cole_bilingue no validas -> nan
for i in range(0, len(arr_cole_delete)):
    pos = arr_cole_delete[i]-i  
    print(pos)
    data_valle = data_valle.drop(data_valle.index[pos])
```
